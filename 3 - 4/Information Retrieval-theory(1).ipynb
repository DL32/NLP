## Regular Expressions

Defining REs in Python is straightforward:

```python
with open ('../data/tweets_en.txt', encoding = 'utf8') as tweet_file:
    tweets = [line.strip() for line in tweet_file]

#tweets[:2]

import re

pattern = re.compile('@[a-zA-Z0-9_\.-]+')

no_name_tweets = [re.sub(pattern, '@USER', _i) for _i in tweets]
no_name_tweets

need_for = re.compile('#[A-Za-z\._)*good[A-Za-z0-9\._]*')

for _i in tweets:
    result = re.search(need_for, _i)
    if result is not None:
        print(result.group())
        
import re

pattern = re.compile('[bcr]at')
```

We can then use the pattern to `search()` or `match()` strings to it. 

`search()` will return a result if the pattern occurs **anywhere** in the input string.

`match()` will only return a result if the pattern matches the **beginning** of the input string.

```python
word = 'the batter won the game'
matches = re.match(pattern, word) # won't return a a result, i.e., matches = None
searches = re.search(pattern, word) # finds a substring
```

Both have a number of attributes to access the results. 
- `span()` gives us a tuple of the substring that matches
- `group()`returns the matched substring

```python
span = searches.span()
word[span[0]:span[1]], span
```

```python
searches.group()
```

If we have used several RE groups (in brackets `()`), we can access them individually via `groups()`

```python
word = 'preconstitutionalism'
affixes = re.compile('(...).+(...)')
re.search(affixes, word).groups()
```

For the email address finder, we can use a more advanced pattern and test it:

```python
# define email pattern
email = re.compile('^[A-Za-z0-9][A-Za-z0-9\.-]*@[A-Za-z0-9][A-Za-z0-9\.-]+\.[A-Za-z0-9\.-][A-Za-z0-9\.-][A-Za-z0-9\.-]?$')

# these should all fail
for address in ['me.@unibocconi.it', '@web.de', '.@gmx.com', 'not working@aol.com']:
    print(re.match(email, address))
bocconi_address = re.match(email, 'me.@unibocconi.it')
```

We can also use the pattern to replace elements of a string that match with `sub()`

```python
numbers = re.compile('[0-9]')
re.sub(numbers, '0', 'Back in the 90s, when I was a 12-year-old, a CD cost just 15,99EUR!')
```

## PMI
Extracting collocations via PMI and other methods from text is relatively straightforward, and `nltk` offers some functions to do so flexibly.

```python
import nltk
# If you are missing the data, you can download it with the next line
#nltk.download('all')
```

First, we need to get some data:

```python
documents = [line.strip() for line in open('../data/moby_dick.txt', encoding='utf8').readlines()]
print(documents[1])
```

Now, we can set up the stopwords and extract a flat list of all words in the corpus:

```python
from nltk.corpus import stopwords

# put predefined stopwords in a set for faster lookup
stopwords_ = set(stopwords.words('english'))

words = [word.lower() for document in documents for word in document.split() 
         if len(word) > 2 
         and word not in stopwords_]
```

Now, we can define the n-gram window and metric to find collocations

```python
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
from collections import Counter

finder = BigramCollocationFinder.from_words(words)
bgm = BigramAssocMeasures()
score = bgm.mi_like
```

And finally, we collect and display the collocations:

```python
collocations = {'_'.join(bigram): pmi 
                for bigram, pmi in finder.score_ngrams(score)}
Counter(collocations).most_common(10)
```
